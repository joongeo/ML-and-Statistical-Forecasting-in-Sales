```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

train = pd.read_csv("../data/train.csv")
calendar = pd.read_csv("../data/calendar.csv")
prices = pd.read_csv("../data/prices.csv")
```

## Data Cleaning

```python
# Delete 'd' in column titles
_cols = list(train.columns)
train.columns = pd.Index(_cols[:6] + [int(c.replace("d_","")) for c in _cols[6:]])
train.columns = train.columns.astype(str)

# Pivot train to long
df_melt = pd.melt(train, id_vars = [i for i in train.columns if i.find("id") != -1],
                          value_vars = [i for i in train.columns if i.isnumeric()], var_name = 'd', value_name = 'sales')

# Remove 'd' in calendar
calendar['d'] = calendar['d'].str[2:]
calendar['date'] = pd.to_datetime(calendar['date'], format='%Y-%m-%d')

# Merge pivoted train with calendar
traincal = pd.merge(df_melt, calendar, on = 'd', how = 'left')
traincal['date'] = pd.to_datetime(traincal['date'], format='%Y-%m-%d')
traincal.d = traincal.d.astype('int')
traincal.head()
```

```python
# Get the total sales by store_id
temp_df = traincal[["date", "sales", "store_id"]]
total_sales_by_store_id = temp_df.groupby(["store_id","date"]).sum().reset_index()
total_sales_by_store_id = total_sales_by_store_id.rename(columns={'sales': 'total_sales_in_cluster',
                                                                  'store_id':'cluster'})

total_sales_by_store_id
```

## Clustering by Store ID

### Feature Creation

In this part of the notebook, we try to create new features based from existing ones. We based it from the suggested features in the slides. Firstly and most obviously, we created day, month, year, and day of week features from the original date. We also created a weekend feature, because sales might be increased on weekends. We also created a holiday feature, which denotes if the given day is a holiday. Based from the description of the task, we assumed that the dataset is from a Singapore's stores sales. Hence, we decided to use Singapore's holidays. We also created lag features for the data, and we found that including one-day, two-day, and three-day lags would be best if the interval between each point is just a day. 

```python
import holidays

# Create an instance of the Singapore holiday class
sg_holidays = holidays.CountryHoliday('SG', years=range(2011, 2017))

# Get the holiday dates as a set
holiday_dates = set(sg_holidays.keys())
```

```python
# Create new features from existing ones
total_sales_by_store_id['day'] = total_sales_by_store_id['date'].dt.day
total_sales_by_store_id['month'] = total_sales_by_store_id['date'].dt.month
total_sales_by_store_id['year'] = total_sales_by_store_id['date'].dt.year
total_sales_by_store_id['day_of_week'] = total_sales_by_store_id['date'].dt.dayofweek
total_sales_by_store_id['is_weekend'] = total_sales_by_store_id['day_of_week'].isin([5, 6])

# Without calling the Timestamp.date function on each of the np.date objects we cannot check for presence or absence in another set
for index, row in total_sales_by_store_id.iterrows():
        total_sales_by_store_id.at[index, 'is_holiday'] = pd.Timestamp.date(row['date']) in holiday_dates
        
total_sales_by_store_id['is_holiday'] = total_sales_by_store_id['is_holiday'].astype(np.bool_)

# Creating lag features
# Sales for previous 1, 2 and 3 days for each date. 
total_sales_by_store_id['sales_lag1'] = total_sales_by_store_id['total_sales_in_cluster'].shift(1)
total_sales_by_store_id['sales_lag2'] = total_sales_by_store_id['total_sales_in_cluster'].shift(2)
total_sales_by_store_id['sales_lag3'] = total_sales_by_store_id['total_sales_in_cluster'].shift(3)

total_sales_by_store_id.head()
```

### Forecasting

We used two distance measures to cluster our time series data for use in the XGBoost model: static weekly proportions and weekly proportions using a rolling window. In the static distance measure, the distance from an item to its cluster centroid is the proportion of its sales compared to the total sales made by the cluster. We fix our proportions based on the total sales of the last week in our training dataset. In the rolling window approach, we update our proportion value every time a prediction is made such that our new proportion value is inclusive of our last prediction. Each window represents the week before the day in question, and the window moves after we forecast one day. These distance measures help to group similar time series together and improve the accuracy of our forecasts.

```python
# Define our forecasting functions
import xgboost as xgb
from math import sqrt
from sklearn.metrics import mean_squared_error

def xgboost_cluster(df, cluster):
    """
    df has the columns: 'cluster', 'date', 'sales', etc.
    cluster is the cluster of interest
    """
    
    df = df[df['cluster'] == cluster]
    
    # Define our features
    features = ['day', 'month', 'year', 'day_of_week', 'is_weekend', 
                'is_holiday', 'sales_lag1', 'sales_lag2', 'sales_lag3']
    target = 'total_sales_in_cluster'
    
    # Set X and y
    X = df[features]
    y = df[target]
        
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    model = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                             n_estimators=3000,
                             max_depth=5,
                             early_stopping_rounds=50,
                             learning_rate=0.005,
                             reg_lambda=0.5,
                             reg_alpha=0.5)
    
    model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], 
              verbose=False)
    
    test_pred = model.predict(X_test) # For obtaining metrics values
    rmse = sqrt(mean_squared_error(y_test, test_pred))
    mae = abs(y_test - test_pred).mean()
    mape = abs((y_test - test_pred) / y_test).mean()
    prediction = model.predict(X_test[-21:]) # Predicts last 21 days
    return prediction, rmse, mae, mape

# For loop to run auto arima for all unique store_id and subcat_id
def forecast_cluster(df, csvdest):
    """
    df has the columns: 'cluster', 'date', 'sales'
    csvdest is the file name for the prediction csv in string
    """
    prediction = pd.DataFrame()
    clusters = df['cluster'].unique()
    rmses = []
    maes = []
    mapes = []
    for cluster in clusters:
        predict, rmse, mae, mape = xgboost_cluster(df, cluster)
        rmses.append(rmse)
        maes.append(mae)
        mapes.append(mape)
        prediction[cluster] = predict
        prediction.to_csv(f"{csvdest}.csv")
        print(f"Cluster: {cluster} is predicted")
    return rmses, maes, mapes

def distance1(df, days = np.array(range(1920, 1941))):
    """
    df has id, cluster, proportion, and the prediction days as the columns
    """
    for d in days:
        df[d] = df[d] * df['prop'] #columnwise multiplication # Removed round()
        
def rolling_window(train_df, prediction_df):
    """
    train_df is traincal with the natural cluster of interest column name renamed to be 'cluster'
    prediction_df has cluster name and prediction days as the columns
    """
    prediction = pd.DataFrame()
    train_temp = train_df[['id', 'd', 'cluster', 'sales']]
    train_temp['d'] = train_temp['d'].astype(int)
    
    for day in np.array(range(1920, 1941)):
        temp = train_temp[train_temp['d'] >= day-7]
        
        total_cluster_sales = temp.groupby(["cluster","d"]).sum().reset_index().rename(columns={'sales':'total_cluster_sales'})
    
        merged = pd.merge(temp, total_cluster_sales, on = ['cluster', 'd'], how = 'left')
        #get weekly sums for item and stores
        merged['item_weekly_sales'] = pd.DataFrame(merged.groupby('id')['sales'].transform('sum'))
        merged['cluster_weekly_sales'] = pd.DataFrame(merged.groupby(['cluster', 'id'])['total_cluster_sales'].transform('sum'))
        #set distance
        merged['prop'] = merged['item_weekly_sales'] / merged['cluster_weekly_sales']
        
        merged_temp_subset = merged[['id', 'cluster', 'prop']]
        
        forecast_df = pd.merge(merged_temp_subset, prediction_df, on = 'cluster', how ='left')
        forecast_df[day] = round(forecast_df['prop'] * forecast_df[day])
        preds_for_day = forecast_df[['id', 'cluster', day]].drop_duplicates().rename(columns={day:'sales'})
        preds_for_day['d'] = [day]*30490
        
        train_temp = train_temp.append(preds_for_day)
        prediction[day] = preds_for_day[['sales']]
        print(f"Prediction made for day: {day}")
        
    return prediction 
```

### Analysis
As expected, our rolling window forecasts using the XGBoost model scored lower than our fixed proportion forecasts. This is because our rolling window approach includes our previous forecasts, which are not always perfectly accurate, and can produce distorted distances from an item to its cluster. The rolling window distance measure was still effective in grouping similar time series together, but the added noise from the previous forecasts led to slightly lower forecast accuracy compared to the static weekly proportion approach.

## Summary

In summary, in our data preparation stage, we did not use transformations such as boxcox as we found that it decreased the performance of our models significantly. We trained two models using two natural clusters, subcategory id and store id, on their respective cluster centroids defined as the total sales per day. To obtain the predictions for each item, we used XGBoost and trained the model on several features, such as day, month, year, whether or not a date was a weekend and/or a holiday, and one-day, two-day, and three-day lags. We also used both the static weekly proportion and rolling window distance measures on both clusters. 

We found that the model trained on the subcategory id cluster without the rolling window distance measure performed the best, because it achieved the highest accuracy and lowest error rates in predicting the sales of items across multiple stores. The rolling window distance measure introduced too much noise and resulted in less accurate predictions. Additionally, the static weekly proportion method allowed the model to capture seasonal trends and sales patterns, which was crucial in accurately predicting sales for the upcoming weeks. Overall, our data preparation and modeling approach proved to be effective in accurately predicting the sales of items across multiple stores, and can be applied to similar retail datasets.

## Future Directions
We explored using DTW clustering to group items based on their time-series similarity, but ultimately opted to use subcategory and store clusters with XGBoost for our models due to computational limitations. However, we believe that if we had more resources, using DTW clustering could potentially improve our results. In addition, running XGBoost on each item_id could further improve the accuracy of our forecasts. We also recognize the potential value of incorporating insights on anomalies within the general trend, such as spikes and plateaus, for more accurate forecasting.
